```python
import numpy as np
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt

# Fazer upload da imagem
print("Por favor, selecione a imagem para upload.")
uploaded = files.upload()

# Carregar a imagem enviada
for filename in uploaded.keys():
    # Abrir e converter a imagem para um array NumPy
    imagem = Image.open(filename)
    imagem = np.array(imagem)

# Função para converter para tons de cinza
def converter_para_cinza(imagem):
    if len(imagem.shape) == 3:  # Verifica se é colorida (RGB)
        return (0.2989 * imagem[:, :, 0] + 0.5870 * imagem[:, :, 1] + 0.1140 * imagem[:, :, 2]).astype(np.uint8)
    return imagem

# Função para binarizar a imagem (branco e preto)
def binarizar(imagem, limite=127):
    return np.where(imagem > limite, 255, 0).astype(np.uint8)

# Transformações
imagem_cinza = converter_para_cinza(imagem)
imagem_binaria = binarizar(imagem_cinza)

# Visualizar as imagens lado a lado
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
titulos = ['Original (Colorida)', 'Tons de Cinza', 'Binária (B/W)']
imagens = [imagem, imagem_cinza, imagem_binaria]

for ax, img, titulo in zip(axes, imagens, titulos):
    if len(img.shape) == 2:  # Tons de cinza ou binária
        ax.imshow(img, cmap='gray')
    else:  # Colorida
        ax.imshow(img)
    ax.set_title(titulo)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
